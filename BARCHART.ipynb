```python
# Visualize Categorical Variables:
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("heart_disease_health_indicators_BRFSS2015 2 (1).csv")

# Define the categorical columns to visualize
categorical_columns = ['Smoker', 'PhysActivity', 'Fruits', 'Veggies']

# Set up the matplotlib figure
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Define a custom color palette
custom_palette = sns.color_palette("Set2")

# Plot bar charts for each categorical variable
for ax, col in zip(axes.flatten(), categorical_columns):
    sns.countplot(x=col, data=df, ax=ax, palette=custom_palette)
    ax.set_title(f'Distribution of {col}')
    ax.set_xlabel(col)
    ax.set_ylabel('Count')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
#Compare Heart Disease Across Categories:
# Set up the matplotlib figure
fig, axes = plt.subplots(2, 2, figsize=(14, 12))

# Plot grouped bar charts to compare the prevalence of heart disease across different categorical variables
for ax, col in zip(axes.flatten(), categorical_columns):
    grouped_data = df.groupby([col, 'HeartDiseaseorAttack']).size().unstack().fillna(0)
    grouped_data.plot(kind='bar', stacked=True, ax=ax)
    ax.set_title(f'Heart Disease Prevalence by {col}')
    ax.set_xlabel(col)
    ax.set_ylabel('Count')

# Adjust layout
plt.tight_layout()
plt.show()
```
